## stack / unstack
- index를 기준으로 데이터를 피봇하는 함수
- stack: 각 행의 가로 형태 Series를 세로 형태의 Series로 변환. 컬럼을 2nd index(level 1)로 추가하여 multiindex dataframe을 반환
- unstack: stack의 반대. 
    - Multiindex의 Dataframe의 경우, 특정 index을 기준으로 세로 형태의 Series를 가로 형태로 변환하여, 해당 index의 데이터가 컬럼으로 변환
    - (사용X) Single index의 경우, 해당 index가 2nd index로 설정되고 컬럼이 1st index로 설정되어 multiindex dataframe가 반환됨

## pivot / pivot_table
- pivot: 데이터 구조를 피봇. unstack을 사용하는 것과 동일. multiindex를 설정할 수 없음
- pivot_table: pivot과 동일한 기능 + values를 집계하는 기능. multiindex 설정 가능

```python
snp_stocks = snp_stocks.reset_index()
```

```python
snp_stocks.pivot(index='Date', columns='Symbol', values='Adj Close')
```

```python
snp_comps = snp_comps.reset_index()
```

```python
snp_comps.pivot_table(index='Country', columns='Sector', values='Weight')
```

```python
snp_comps.pivot_table(index='Country', columns='Sector', values='Weight', aggfunc='sum')
```

```python
snp_comps.pivot_table(index=['Country', 'Exchange'], columns='Sector', values=['Weight', 'Marketcap'])
```

## squeeze

- 인풋값에 대한 구조를 모를떄(DataFrame인지 Serires인지 모를때), 해당 값을 하나 아래 tpye으로 변환하는 함수

```python
snp_comps[['Symbol']].squeeze()
```
